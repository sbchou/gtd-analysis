```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas
import inspect
from datetime import datetime
import seaborn as sns
import glob
from os.path import basename
sns.set_context('poster')
import re
import csv
```

```python
df_hist = pandas.read_excel('../data/globalterrorismdb_0616dist.xlsx')
```

```python
df_16 = pandas.read_excel('../data/Global Terrorism Database 2015-2016 Preliminary Data EMBARGOED.xlsx')
df_16 = df_16[df_16.iyear == 2016]
```

```python
df = pandas.concat([df_hist, df_16]) 
```

```python
print df.iyear.min()
print df.iyear.max()
```

```python
df.to_csv('../data/MASTER-1970-2016.csv', encoding='utf8')
```

```python
dupes = df[df.related.notnull() & (df.iyear >= 1997)] 
dupes['related_ls'] = dupes.related.apply(lambda x: [y.strip() for y in x.split(",")])
```

```python
l1 = list(dupes.related_ls)
l2 = list(dupes.eventid.apply(str).apply(unicode))
```

```python
print len(l1)
print len(l2)
```

```python
master_ls = [x + [y] for x,y in zip(l1, l2)]
```

```python
master_ls
```

```python
sorted_master_ls = [sorted(l) for l in master_ls]
```

```python
import csv

with open("../data/all-dupe-groups-after-1997.csv", "wb") as f:
    writer = csv.writer(f)
    writer.writerows(sorted_master_ls)
```

```python
ids_to_keep = pandas.read_excel('../data/first-of-dupes.xlsx', index_col=0)
```

```python
all_dupe_ids = list(dupes.eventid)
```

```python
ids_to_keep = list(ids_to_keep.index)
```

```python
ids_to_delete = set(all_dupe_ids) - set(ids_to_keep)
```

```python
criterion = lambda row: row['eventid'] not in ids_to_delete
deduped = df[df.apply(criterion, axis=1)]
```

```python
deduped = deduped[deduped.iyear >= 1997]
```

```python
deduped.sort_values('eventid').to_csv('../data/1997-2016-DEDUPED.csv', encoding='utf8')
```
